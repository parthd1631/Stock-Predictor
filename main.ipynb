```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index, name = "Predictions")
    combined = pd.concat([test["Target"], preds], axis = 1)
    return combined
```

```python
def backtest(data, model, predictors, start = 2500, step = 250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)

    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500, model, predictors)
```

```python
def get_stock_data_yq(ticker):
    try:
        stock = Ticker(ticker)
        data = stock.history(period="5y")
        if not isinstance(data, pd.DataFrame):
            return ticker, None
        if len(data) > 0:
            # Reset multi-index to make symbol a column
            data = data.reset_index()
            data = data[data['symbol'] == ticker]
            data = data.set_index('date')
            
            # Convert price columns to numeric
            for col in ['open', 'high', 'low', 'close', 'volume']:
                if col in data.columns:
                    data[col] = pd.to_numeric(data[col], errors='coerce')
            
            data["Tomorrow"] = data["close"].shift(-1)
            data["Target"] = (data["Tomorrow"] > data["close"]).astype(int)
            
            # Rename columns to match yfinance format
            data = data.rename(columns={
                'open': 'Open', 
                'high': 'High', 
                'low': 'Low', 
                'close': 'Close', 
                'volume': 'Volume'
            })
            return ticker, data
        return ticker, None
    except Exception as e:
        print(f"Error with {ticker}: {e}")
        return ticker, None

def get_all_stocks_data_batch(tickers):
    stock_data = {}
    # Process tickers in batches of 100
    batch_size = 100
    for i in range(0, len(tickers), batch_size):
        batch = tickers[i:i+batch_size]
        stocks = Ticker(batch)
        data = stocks.history(period="5y")
        if isinstance(data, pd.DataFrame):
            for ticker in batch:
                try:
                    ticker_data = data.reset_index()
                    ticker_data = ticker_data[ticker_data['symbol'] == ticker]
                    if len(ticker_data) > 0:
                        ticker_data = ticker_data.set_index('date')
                        
                        # Convert price columns to numeric
                        for col in ['open', 'high', 'low', 'close', 'volume']:
                            if col in ticker_data.columns:
                                ticker_data[col] = pd.to_numeric(ticker_data[col], errors='coerce')
                        
                        ticker_data["Tomorrow"] = ticker_data["close"].shift(-1)
                        ticker_data["Target"] = (ticker_data["Tomorrow"] > ticker_data["close"]).astype(int)
                        
                        # Rename columns to match yfinance format
                        ticker_data = ticker_data.rename(columns={
                            'open': 'Open', 
                            'high': 'High', 
                            'low': 'Low', 
                            'close': 'Close', 
                            'volume': 'Volume'
                        })
                        stock_data[ticker] = ticker_data
                except Exception as e:
                    print(f"Error processing {ticker}: {e}")
    return stock_data
```

```python
# Get a list of tickers (example: using S&P 500 components)
def get_tickers():
    # Option 1: From Wikipedia
    table = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
    df = table[0]
    return df['Symbol'].tolist()
    
    # Option 2: You can use any custom list
    # return ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'META']

# Get the tickers
tickers = get_tickers()
print(f"Found {len(tickers)} tickers")

# Fetch data for all tickers using batch processing
stock_data = get_all_stocks_data_batch(tickers)
print(f"Successfully fetched data for {len(stock_data)} stocks")
```

```python
# Add features for each stock's dataframe
def add_features(df):
    # Ensure columns are numeric types first
    numeric_columns = ['Open', 'High', 'Low', 'Close', 'Volume', 'Tomorrow', 'Target']
    
    # Check for any non-numeric columns that should be numeric
    for col in numeric_columns:
        if col in df.columns and df[col].dtype == 'object':
            df[col] = pd.to_numeric(df[col], errors='coerce')
    
    # Drop rows with NaN values after conversion
    df = df.dropna(subset=['Close', 'Tomorrow'])
    
    # Now proceed with feature creation
    horizons = [2, 5, 60, 250]
    new_predictors = []
    
    for horizon in horizons:
        # Only include numeric columns in rolling
        numeric_df = df[['Close']]
        rolling_averages = numeric_df.rolling(horizon).mean()
        
        ratio_column = f"Close_Ratio_{horizon}"
        df[ratio_column] = df["Close"] / rolling_averages["Close"]
        new_predictors.append(ratio_column)
        
        # Ensure Target is numeric
        if df['Target'].dtype == 'object':
            df['Target'] = pd.to_numeric(df['Target'], errors='coerce')
        
        # Calculate trend using Target column
        trend_column = f"Trend_{horizon}"
        df[trend_column] = df['Target'].shift(1).rolling(horizon).sum()
        new_predictors.append(trend_column)
    
    return df.dropna(), new_predictors
# Train and predict for each stock
def analyze_all_stocks(stock_data):
    results = {}
    
    for ticker, df in stock_data.items():
        try:
            # Skip tickers with insufficient data
            if len(df) < 1000:  # Adjust this threshold as needed
                continue
                
            # Add features
            df, predictors = add_features(df)
            if df.empty:
                continue
                
            # Create model
            model = RandomForestClassifier(n_estimators=200, min_samples_split=50, random_state=1)
            
            # Train and test
            train = df.iloc[:-100]
            test = df.iloc[-100:]
            
            if len(train) < 500:  # Skip if not enough training data
                continue
                
            model.fit(train[predictors], train["Target"])
            
            preds = model.predict_proba(test[predictors])[:, 1]
            preds[preds >= 0.6] = 1
            preds[preds < 0.6] = 0
            
            # Calculate precision
            precision = precision_score(test["Target"], preds, zero_division=0) if len(test) > 0 else None
            results[ticker] = {
                'precision': precision,
                'predictions': preds,
                'model': model,
                'data': df
            }
            
        except Exception as e:
            print(f"Error analyzing {ticker}: {e}")
            
    return results

# Run the analysis
results = analyze_all_stocks(stock_data)

# Display results sorted by precision
precision_scores = {ticker: data['precision'] for ticker, data in results.items() if data['precision'] is not None}
sorted_scores = sorted(precision_scores.items(), key=lambda x: x[1], reverse=True)

print("Top 10 stocks by prediction precision:")
for ticker, score in sorted_scores[:10]:
    print(f"{ticker}: {score:.4f}")
```
